```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats
```

```python
# read data
fish= pd.read_csv("./data/2-2-1-fish.csv")
```

```python
#3行目まで表示
fish.head(3)
```

```python
# ヒストグラムの作成
plt.hist(fish['length'])
```

```python
fish_10=fish.loc[0:10]
```

```python
#ラグプロット
sns.rugplot(fish_10['length'])
```

```python
sns.kdeplot(fish['length'])
```

```python
# バンド幅を変える
plt.figure(figsize=(10, 4))
ax=sns.kdeplot(fish['length'],bw=1,label=1)
ax=sns.kdeplot(fish['length'],ax=ax,bw=0.25,label=0.25)
ax=sns.kdeplot(fish['length'],ax=ax,bw=4,label=4)
ax.set(xlim = (2,18))
```

```python
# 平均
print(fish.mean())
```

```python
# 等差数列
suuretu=np.array(range(1001))
print(suuretu)
# 数列の長さ
print(len(suuretu))
# 中央値
print(np.median(suuretu))
# パーセンタイル(50%)
print(np.percentile(suuretu,50))
# パーセンタイル(25%)
print(np.percentile(suuretu,25))
# パーセンタイル(75%)
print(np.percentile(suuretu,75))
```

```python
from statsmodels import datasets
import statsmodels.api as sm

nile=datasets.get_rdataset('Nile',cache=True).data
nile_values=nile['value']
fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(111)
sm.graphics.tsa.plot_acf(nile_values, lags=20, ax=ax1) 
plt.show()
```
