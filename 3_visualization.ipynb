```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats
from scipy.stats import gaussian_kde
```

## 3-3 read data

```python
# read data
fish_df= pd.read_csv("./data/2-2-1-fish.csv")
fish_np=fish_df.values
```

## 3-4 histgram and kernel density

```python
# data frame based
fish_df["length"].plot(kind="hist", bins=20, alpha=0.5)
plt.xlim(3,16)
plt.show()
```

```python
fish_df["length"].plot(kind="kde", secondary_y=True)
plt.xlim(3,16)
plt.show()
```

## 3-5 two graph  visualization

```python
# data frame based
fish_df["length"].plot(kind="hist", bins=20, alpha=0.5)
fish_df["length"].plot(kind="kde", secondary_y=True)
plt.xlim(3,16)
plt.show()
```
